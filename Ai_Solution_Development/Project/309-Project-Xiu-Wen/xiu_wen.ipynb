## 3. Dataset Merging: Build a Master Customer Table

Merge pathway:

orders_dataset as the main transaction table. 

Join: 
1) customers_dataset on customer_id
2) geolocation_dataset on customer_zip_code_prefix
3) order_items_dataset on order_id
4) order_payments_dataset on order_id
5) order_reviews_dataset on order_id
6) products_dataset on product_id
7) sellers_dataset on geolocation_zip_code_prefix
8) product_category_name_translation_dataset on product_category_name 

```python
pd.DataFrame({'x': [1, 2, 3], 'name': ['Alice', 'Bob', 'Charlie']})
order_items_dataset,customers_dataset.columns, geolocation_dataset.columns, order_items_dataset.columns, order_payments_dataset.columns, order_reviews_dataset.columns, orders_dataset.columns, products_dataset.columns, sellers_dataset.columns, product_category_name_translation_dataset.columns        
```

```python
# customers_dataset['customer_zip_code_prefix']
customers_dataset
```

```python
# geolocation_dataset['geolocation_zip_code_prefix']
geolocation_dataset
```

```python
customers_dataset.rename(columns={'customer_zip_code_prefix': 'zip_code_prefix'}, inplace=True)
geolocation_dataset.rename(columns={'geolocation_zip_code_prefix': 'zip_code_prefix'}, inplace=True)
```

```python
sellers_dataset.rename(columns={'seller_zip_code_prefix': 'zip_code_prefix'}, inplace=True)
sellers_dataset
```

```python
# Merge orders_dataset with customers_dataset. 
master_dataset = pd.merge(orders_dataset, customers_dataset, on='customer_id', how='left')

# Merge with geolocation_dataset.
master_dataset = pd.merge(master_dataset, geolocation_dataset, on='zip_code_prefix', how='left')

# Merge with order_items_dataset.
master_dataset = pd.merge(master_dataset, order_items_dataset, on='order_id', how='left')

# Merge with order_payments_dataset.
master_dataset = pd.merge(master_dataset, order_payments_dataset, on='order_id', how='left')

# Merge with order_reviews_dataset.
master_dataset = pd.merge(master_dataset, order_reviews_dataset, on='order_id', how='left')

# Merge with products_dataset.
master_dataset = pd.merge(master_dataset, products_dataset, on='product_id', how='left')

# Merge with sellers_dataset.
master_dataset = pd.merge(master_dataset, sellers_dataset, on='zip_code_prefix', how='left')

# Merge with product_category_name_translation_dataset.
master_dataset = pd.merge(master_dataset, product_category_name_translation_dataset, on='product_category_name', how='left')
```
